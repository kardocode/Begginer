```python
import pandas as pd
```

```python
ride_sharing = pd.read_csv('https://raw.githubusercontent.com/pradhrahul/Data-Science-Visualization/refs/heads/main/ride_sharing_clean.csv')
```

```python
ride_sharing.info()
```

```python
ride_sharing.head()
```

```python
ride_sharing['user_type'] = ride_sharing['user_type'].astype(int)
ride_sharing['user_type'].info()
```

```python
ride_sharing['user_type'].describe()
```

```python
ride_sharing['user_type'].astype('category')
```

```python
ride_sharing['duration'].dtype
```

```python
duration_trim = ride_sharing['duration'].str.strip("minutes")
```

```python
ride_sharing['duration'] = duration_trim.astype('int')
```

```python
my_dict = {1:"for free ride",2:"for pay ride",3:"for monthly subscribe"}
```

```python
ride_sharing["new_user_type"] = ride_sharing['user_type'].map(my_dict)
```

```python
ride_sharing['new_user_type'].describe()
```

```python
ride_sharing['duration'].mean()
```

```python
ride_sharing['tire_sizes'] = ride_sharing['tire_sizes'].astype(int)
```

```python
ride_sharing['tire_sizes'].dtype
```

```python
ride_sharing['tire_sizes'].head()
```

```python
#Use .loc[] to set all values of tire_sizes above 27 to 27.
ride_sharing["tire_sizes"] = ride_sharing["tire_sizes"].astype(int)
ride_sharing.loc[ride_sharing["tire_sizes"] > 27, "tire_sizes"] = 27
```

```python
ride_sharing['tire_sizes'] = ride_sharing['tire_sizes'].astype('category')
ride_sharing['tire_sizes'].describe()
```

```python
ride_sharing["ride_dt"] = pd.to_datetime(ride_sharing["ride_dt"]).dt.date
```

```python
from datetime import date
today = date.today()
print(today)
```

```python

ride_sharing.loc[ride_sharing["ride_dt"] > today, "ride_dt"] = today

ride_sharing['ride_dt'].describe()
```

```python
ride_sharing['ride_dt'].max()
```

```python
#Find duplicated rows of ride_id in the ride_sharing DataFrame while setting keep to False.
dupes = ride_sharing.duplicated(subset = 'ride_id', keep = False)
print(duplicates)
```

```python
duplicated_rides = dupes.sort_values()
```

```python
#Create the statistics dictionary which holds minimum aggregation for user_birth_year and mean aggregation for duration.
statistics = {'user_birth_year': 'min', 'duration': 'mean'}
```

```python
#Drop incomplete duplicates by grouping by ride_id and applying the aggregation in statistics.
ride_unique = ride_sharing.groupby('ride_id').agg(statistics).reset_index()
ride_unique
```

```python
#Find duplicates again and run the assert statement to verify de-duplication.
duplicates = ride_unique.duplicated(subset = 'ride_id', keep = False)
duplicated_rides = ride_unique[duplicates == True]
```
