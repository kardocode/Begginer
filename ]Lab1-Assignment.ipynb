#Grumpy Wizard's Number Sorter
You are a junior apprentice to a very, very grumpy wizard named Bartholomew. He has a magical sorting cauldron that can only handle ingredients with numbers that are "just right." If a number is too high, it explodes. If it's too low, it turns into a frog. If it's zero, nothing happens, which he finds terribly boring.

To avoid a grumpy Bartholomew and a frog-infested lab, your task is to write a script that takes a number as input and determines its category for the cauldron:

* **Positive**: The number is "Too High!" and will cause an explosion.
* **Negative**: The number is "Too Low!" and will turn into a frog.
* **Zero**: The number is "So Boring." and will disappoint Bartholomew.

Your script will be the key to maintaining a peaceful (and frog-free) wizard's laboratory!

```python
n = int(input("Enter a number"))
print("Number entered:", n)
if n == 0:
  print(f"Number {n} is Zero")
elif n > 0:
  print(f"Number {n} is Positive")
else:
  print(f"Number {n} is Negative")
```
