```python
import pandas as pd
import numpy as np
```

```python
airline = pd.read_csv('https://raw.githubusercontent.com/pradhrahul/Data-Science-Visualization/refs/heads/main/airlines_survey_data.csv')
```

```python
categories = pd.DataFrame ({
     'cleanliness': ['Clean', 'Average', 'Somewhat clean', 'Somewhat dirty', 'Dirty'],
    'safety': ['Neutral', 'Very safe', 'Somewhat safe', 'Very unsafe', 'Somewhat unsafe'],
    'satisfaction': ['Very satisfied', 'Neutral', 'Somewhat satisfied', 'Somewhat unsatisfied', 'Very unsatisfied']
})
print (categories)
```

```python
print(airline.survey_response.unique())
```

```python
airline.dest_region = airline.dest_region.str.lower()
airline.tail()
```

```python

airline.dest_region = airline.dest_region.replace('eur','europe');
print(airline.dest_region.unique())
```

```python
print(airline.dest_size.unique())
airline.dest_size = airline.dest_size.str.strip(' ')
print(airline.dest_size.unique())
```

```python
day = airline.day;
airline['day_week'] = ['weekday' if x!= 'Sunday' else 'weekend' for x in day];
airline.day_week.unique()
```

```python
label_ranges = [0,60,180,np.inf]
label_names = ['short','medium','long']
airline['wait_type'] = pd.cut(airline['wait_min'], bins=label_ranges, labels=label_names,right = False)
airline.head()
```

```python
dry_mapping = {
    'Monday': 'weekday',
    'Tuesday': 'weekday',
    'Wednesday': 'weekday',
    'Thursday': 'weekday',
    'Friday': 'weekday',
    'Saturday': 'weekday',
    'Sunday': 'weekend',
}
airline['day_week'] = airline.day.replace(dry_mapping)
airline.head()
```

```python
pattern = ["Dr.","Mr.","Miss","Ms."];
for i in pattern:
  airline.full_name = airline["full_name"].str.replace(i,'',regex=False);
airline.head()
```
