```python
# Regular Expression
regex = r'^(\d{4})'

#r -- raw String
#() -- group
#\d -- digit 0 to 9
#{4} -- \d repeats 4 times
```

```python
df['Date of Publication'].str.extract(regex,expand=False)
```

```python
df["Date of Publication"] = pd.to_numeric(df['Date of Publication'].str.extract(regex,expand=False))
```
